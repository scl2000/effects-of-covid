```python
#convert time period to numeric

#filter dates - Q1-2020 through Q3-2021 (only keep quarters 1-7)

start_year = 2020
start_quarter = 1


def convert_time_period(time_period):
    year, quarter = time_period.split('-Q')
    num_quarters = (int(year) - start_year) * 4 + int(quarter) - start_quarter
    return num_quarters

gdp['num_quarters'] = gdp['TIME'].apply(convert_time_period)

gdp_filtered = gdp[gdp['num_quarters'] < 8]
gdp_filtered
```

```python
unique_counties
```

```python
#filter countries of interest

gdp = gdp_filtered[gdp_filtered['Country'].isin(unique_counties)]
gdp = gdp.rename(columns={'Value': 'GDP per capita'})
gdp = gdp[gdp['Measure'] == 'Per Head, US dollars, volume estimates, fixed PPPs, OECD reference year, seasonally adjusted']
```

```python
gdp = gdp.sort_values(['Country', 'num_quarters'])

gdp['gdp_percent_change'] = (gdp['GDP per capita'] - gdp['GDP per capita'].shift(1)) / gdp['GDP per capita'].shift(1) * 100

gdp = gdp.dropna(subset=['gdp_percent_change'])

gdp.head()

```

```python
#POISSON

import pandas as pd
import statsmodels.api as sm


for country in df['country'].unique():
    country_data = df[df['country'] == country]

    model = sm.GLM(country_data['deaths_per100k'], country_data[['num_days']], family=sm.families.Poisson())
    result = model.fit()

    print('Results for {}:'.format(country))
    print(result.summary())
```

```python
#POISSON with fixed effects

import pandas as pd
import statsmodels.api as sm

country_dummies = pd.get_dummies(df['country'], drop_first=True)

model = sm.GLM(df['deaths_per100k'], sm.add_constant(country_dummies.join(df[['num_days']])), family=sm.families.Poisson())
result = model.fit()

print(result.summary())

```

```python
##DEATH OUTPUT

model = sm.GLM(df['deaths_per100k'], sm.add_constant(country_dummies.join(df[['num_days']])), family=sm.families.Poisson())
result = model.fit()

country_coefs = [coef for coef in result.params.index if any(country in coef for country in unique_countries)]

summary_table = result.summary2()
coefs = summary_table.tables[1].loc[country_coefs, 'Coef.']
p_values = summary_table.tables[1].loc[country_coefs, 'P>|z|']

ranks = coefs.rank(ascending=True)

deathcount_df = pd.DataFrame({'Deaths Coefficient': coefs, 'Deaths P-value': p_values, 'Death count ranks': ranks})

deathcount_df = deathcount_df.sort_values(by='Death count ranks')

print(deathcount_df)
```

```python
#OLS FOR TOTAL DEATHS PER 100K
predictor_vars = ['numday']

mtable = pd.DataFrame(columns=['Country', 'Intercept', 'Slope_numday', 'R-squared', 'p-value'])

for country in df['country'].unique():

    country_data = df[df['country'] == country]
    country_data = country_data.groupby('country').transform(lambda x: x.fillna(x.mean()))

    X = sm.add_constant(country_data[predictor_vars])
    y = country_data['total_deathsper100k']
    model = sm.OLS(y, X).fit()

    intercept, slope_numday = model.params
    rsquared = model.rsquared
    pvalue = model.f_pvalue
    mtable = pd.concat([mtable, pd.DataFrame({'Country': country, 'Intercept': intercept, 'Slope_numday': slope_numday, 'R-squared': rsquared, 'p-value': pvalue}, index=[0])], ignore_index=True)

mtable['ranks'] = mtable['Slope_numday'].rank(ascending=True)

mtable = mtable.sort_values('ranks')

mtable = mtable.reset_index(drop=True)

mtable['ranks'] = mtable['Slope_numday'].rank(ascending=True)

mtable = mtable.sort_values('ranks')

mtable = mtable.reset_index(drop=True)

## KEEPING ONLY CERTAIN COLUMNS

mtable = mtable.loc[:, ['Country', 'Slope_numday', 'R-squared', 'p-value', 'ranks']].rename(columns={'Slope_numday': 'deaths_OLS_coef'})

mtable = mtable.sort_values('ranks')

mtable = mtable.reset_index(drop=True)
deaths_OLS = mtable.rename(index=str)
```

```python
#MV OLS FOR TOTAL DEATHS PER 100K

predictor_vars = ['numday', 'population_density', 'gdp_per_capita', 'diabetes_prevalence']

mtable = pd.DataFrame(columns=['Country', 'Intercept', 'Slope_numday', 'Slope_population_density', 'Slope_gdp_per_capita', 'Slope_diabetes_prevalence', 'R-squared', 'p-value'])

for country in df['country'].unique():

    country_data = df[df['country'] == country]
    country_data = country_data.groupby('country').transform(lambda x: x.fillna(x.mean()))
    X = sm.add_constant(country_data[predictor_vars])
    y = country_data['total_deathsper100k']
    model = sm.OLS(y, X).fit()
    intercept = model.params[0]
    slope_numday, slope_population_density, slope_gdp_per_capita, slope_diabetes_prevalence = model.params
    rsquared = model.rsquared
    pvalue = model.f_pvalue
    mtable = pd.concat([mtable, pd.DataFrame({'Country': country, 'Intercept': intercept, 'Slope_numday': slope_numday, 'Slope_population_density': slope_population_density, 'Slope_gdp_per_capita': slope_gdp_per_capita, 'Slope_diabetes_prevalence': slope_diabetes_prevalence, 'R-squared': rsquared, 'p-value': pvalue}, index=[0])], ignore_index=True)
    
mtable['deaths_MV_OLS_ranks'] = mtable['Slope_numday'].rank(ascending=True)

mtable = mtable.sort_values('deaths_MV_OLS_ranks')
mtable = mtable.reset_index(drop=True)


mtable = mtable.loc[:, ['Country', 'Slope_numday', 'R-squared', 'p-value', 'deaths_MV_OLS_ranks']].rename(columns={'Slope_numday': 'MV_deaths_coef'})

mtable = mtable.sort_values('deaths_MV_OLS_ranks')

mtable = mtable.reset_index(drop=True)
deaths_MV_OLS = mtable.rename(index=str)
deaths_MV_OLS = deaths_MV_OLS.rename(columns={'rank': 'deaths_MV_OLS_rank'})
                                                 
```

The Deviance Pseudo R-squared is a measure of how well a GLM regression model fits the data. It is calculated as the difference between the null deviance and the residual deviance, divided by the null deviance.

The null deviance is a measure of the total variation in the response variable that is explained by the intercept-only model (i.e., a model that includes only the constant term). The residual deviance is a measure of the variation in the response variable that is not explained by the model, after accounting for the effects of the independent variables. The Deviance Pseudo R-squared can be interpreted as the proportion of the total variation in the response variable that is explained by the independent variables in the model, relative to the total variation in the response variable.

Note that the Deviance Pseudo R-squared is not the same as the R-squared value that is commonly used in linear regression models. The R-squared value is based on the sum of squared errors and represents the proportion of variation in the response variable that is explained by the independent variables in the model. The Deviance Pseudo R-squared is based on deviance, which is a more general measure of lack of fit that applies to a wider range of models than the sum of squared errors.
